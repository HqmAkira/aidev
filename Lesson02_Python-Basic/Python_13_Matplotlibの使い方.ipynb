## 複数のグラフを一度に表示する方法

### 1つのグラフ領域に複数のグラフを描く

```python
y1 = df.population # 人口
y2 = df.households # 世帯数

plt.plot(y1, color="b", label="population")
plt.plot(y2, color="r", label="households")
plt.legend()
plt.show()
```

```python
plt.plot(y1, color="b", label="population")
plt.plot(y2, color="r", label="households")
plt.legend(loc="upper left")
plt.show()
```

### 複数のグラフを並べて表示する

#### 横に並べる

##### subplot() を使う

```python
y1 = df.population # 人口
y2 = df.households # 世帯数

# Figureオブジェクトを生成
fig = plt.figure()

# population用の描画領域を作成してグラフを追加
sp1 = fig.add_subplot(1, 2, 1)
sp1.plot(y1)

# households用の描画領域を作成してグラフを追加
sp2 = fig.add_subplot(1, 2, 2)
sp2.plot(y2)

# グラフを表示
plt.show()
```

##### subplots() を使う

```python
fig, axes = plt.subplots(1, 2) # (縦方向のsubplot数, 横方向)

axes[0].plot(y1)
axes[1].plot(y2)

plt.show()
```

#### 縦に並べる

```python
fig, axes = plt.subplots(2, 1) # (縦方向のsubplot数, 横方向)

axes[0].plot(y1)
axes[1].plot(y2)

plt.show()
```

## グラフのサイズを変更する方法

```python
fig = plt.figure(figsize=(10, 4))

sp1 = fig.add_subplot(1, 2, 1)
sp1.plot(y1)

sp2 = fig.add_subplot(1, 2, 2)
sp2.plot(y2)
plt.show()
```

`figsize`引数に長さ2の数値タブル（リストでも良い）を渡すことで、`(横幅インチ, 縦幅インチ)`を指定することが出来ます。

この方法は、最初に扱ったような単一グラフでも使えるので、覚えておきましょう。

```python
x = [7, 8, 11, 14, 17, 21, 22, 26, 29, 30]
y = [19, 36, 43, 54, 55, 64, 88, 90, 92, 97]

plt.figure(figsize=(10, 4))
plt.plot(x, y, ls="--", marker="D", c="r")
plt.show()
```

## 応用例：近似直線の描き方

```python
# サンプルデータ
x = np.array([20, 30, 40, 50, 60, 43, 94, 24, 44])
y = np.array([30, 38, 41, 49, 62, 59, 61, 43, 24])

# 近似直線の式の a と b が入ったndarray 形式のデータを得る
p = np.polyfit(x, y, 1)

# 一次関数の式のオブジェクトを生成する
f = np.poly1d(p)

# 散布図と近似直線を描く
plt.scatter(x, y)
plt.plot(x, f(x), color="black")
plt.show()
```
